```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Создание искусственного набора данных
# Генерация случайных данных
np.random.seed(0)
x = 2 * np.random.rand(100, 1)  # 100 случайных значений от 0 до 2
y = 4 + 3 * x + np.random.randn(100, 1)  # Лицо y = 4 + 3x + случайный шум

# Преобразование в DataFrame для удобства
data = pd.DataFrame(data=np.hstack((x, y)), columns=['x', 'y'])

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(data[['x']], data['y'], test_size=0.2, random_state=0)

# Создание и обучение модели линейной регрессии
model = LinearRegression()
model.fit(X_train, y_train)

# Предсказание на тестовых данных
y_pred = model.predict(X_test)

# Вывод коэффициентов
print(f"Коэффициент (наклон): {model.coef_[0]}")
print(f"Свободный член (смешивание): {model.intercept_}")

# Визуализация результатов
plt.scatter(X_test, y_test, color='black', label='Тестовые данные')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Предсказанная линия регрессии')
plt.title('Линейная Регрессия')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
